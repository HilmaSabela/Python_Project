# Merge, Join, and Concatenate 

```python
# Neutral Join => Value Irirsan dari gabungan 
# Full Outer Join => Value dari irisan, dan diluar irisan (semuanya)
# Left Outer Join => Semua value dari 'Left' dan Irisannya
# Right Outer Join => Semua value dari 'Right' da Irisannya
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"H:\Phyton Tutorial\Pandas Tutorial\LOTR.csv")
df2 = pd.read_csv(r"H:\Phyton Tutorial\Pandas Tutorial\LOTR 2.csv")

df1
```

```python
df2
```

# Merge Function

```python
df1.merge(df2) # Hanya untuk data yang sama dari 2 dataframe
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID') 
# Yang di merge hanya FellowshipID, jadi FirstName nya nggak di merge, jadi ada x dan y
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName']) # FirstName di merge
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross') # Seperti pengali. tabel 1 (per row) dikalikan degan tabel 2
```

# Join Function

```python
# Join tanpa index
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
# Untuk melihat lsuffix / rsuffix , pakai Shift + Tab
```

```python
# Join akan terlihat manfaatnya, jika diatur menjadi index, dan akan bagus pakai index
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
# lsuffix / rsuffix harus ada | Trus bisa ditambah how = '...'

df3
```

# Concatenate

```python
# Concatenate = menggabungkan, menggabungkan dataframe pertama di ATAS dataframe kedua (tergantung penempatan)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join = 'inner')
```

```python
pd.concat([df1, df2], join = 'outer')
```
